# Data Preprocessing

In this task not all data will be used, only data in 2017. So it is necessary to do filtering at the beginning. Besides that there are some players who make team transfers in the NBA transfer market so that there is duplication of player data. Therefore you can use the df.drop_duplicates() syntax to solve this to produce the same output as the trainer. Delete columns that have as many missing values as the entire row of data. Then you can do additional preprocessing if needed or you can immediately process the data. 

```python
# drop unnamed:0
df = df.drop("Unnamed: 0",axis = 1)
```

```python
# slice to year 2017
df_2017 = df[df['Year'] == 2017.0].reset_index(drop=True)
df_2017.head()
```

```python
# check null columns
df_2017.isnull().sum()
```

```python
# drop column blanl & blank2
df_2017  = df_2017.drop(['blanl','blank2'],axis=1)
```

```python
# check duplicate player
df_2017[df_2017['Player'].duplicated()]
```

```python
# delete duplicates and select the last duplicate data
df_2017 = df_2017.drop_duplicates(subset='Player',keep='last')
df_2017.head()
```

```python
df_2017[df_2017['3P%'].isnull()].head()
```

mostly the null data on `3p%` is player on Center position and does not have 3p attempt

```python
df_2017[df_2017['FT%'].isnull()].head()
```

same with `3p%` the player who does not have `ft%` does not attempt any free throws, after the exploration i assume that null likely the player does not attempt some stats like free throws, 3p, shooting, etc

```python
# fill null data with 0
df_2017.fillna(value=0.0,inplace=True)
```

# Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

## Young Player

```python
df_young = df_2017[['Player','Age','Tm']].groupby('Tm').agg({'Age':'min'})['Age']
```

```python
df_young = df_2017[['Player','Age','Tm']][df_2017['Age'].isin(df_young)].sort_values('Age')
```

```python
df_young = df_young.drop_duplicates('Tm',keep='first').set_index('Tm')
df_young
```

## Old Player

```python
df_old = df_2017[['Player','Age','Tm']].groupby('Tm').agg({'Age':'max'})['Age']
```

```python
df_old = df_2017[['Player','Age','Tm']][df_2017['Age'].isin(df_old)].sort_values('Age',ascending=False)
```

```python
df_old = df_old.drop_duplicates('Tm',keep='first').set_index('Tm')
df_old
```

# Which player has the most minutes played (MP) in each position (Pos)? 

```python
df_mp_player = df_2017[['Player','Pos','MP']].groupby('Pos').agg({'MP':'max'})['MP']
```

```python
df_2017[['Player','Pos','MP']][df_2017['MP'].isin(df_mp_player)].sort_values('MP',ascending=False)
```

# Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)? 

```python
df_tm_2017 = df_2017[['TRB%','AST%','STL%','BLK%','Tm']].groupby('Tm').mean()
```

## Best Total Rebound Percentage Team

```python
# Total Rebound Percentage 
df_tm_2017[['TRB%']][df_tm_2017['TRB%'] == df_tm_2017['TRB%'].max()]
```

## Best Assist Percentage Team

```python
# Assist Percentage
df_tm_2017[['AST%']][df_tm_2017['AST%'] == df_tm_2017['AST%'].max()]
```

## Best Steal Percentage Team

```python
# Steal Percentage
df_tm_2017[['STL%']][df_tm_2017['STL%'] == df_tm_2017['STL%'].max()]
```

## Best Block Percentage Team

```python
# Block Percentage
df_tm_2017[['BLK%']][df_tm_2017['BLK%'] == df_tm_2017['BLK%'].max()]
```
